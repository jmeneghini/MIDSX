```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
import os
import mendeleev as pt
from sqlalchemy import create_engine
```

```python
mat_def_df = pd.read_excel('/home/john/Documents/MIDSX/data/reference_set/02-Case 5 Material Definitions.xlsx', header=3)
# drop rows 20 - 25
mat_def_df.drop(mat_def_df.index[20:26], inplace=True)
mat_def_df
```

```python
def atomic_symbol_to_atomic_number(symbol):
    return pt.element(symbol).atomic_number
```

```python
# load sqlite3 database
db_path = '/home/john/Documents/MIDSX/data/data_sources/EPDL/EPDL.db'

engine = create_engine('sqlite:///' + db_path)
```

```python
# create a df of the following form: MaterialID, Name - C5, Density from mat_def_df
# MaterialID, Name - C5, Density
mat_def_den_df = mat_def_df[['Material', 'Density (g/cm3)']]
mat_def_den_df["Material"] = mat_def_df["Material"] + " - C5"
# rename columns
mat_def_den_df.columns = ['Name', 'Density']
```

```python
# create a df of the following form: MaterialID, ElementID, WeightFraction
# MaterialID, ElementID, WeightFraction
mat_comp_df = pd.DataFrame(columns=['MaterialID', 'ElementID', 'WeightFraction'])
element_names_row = mat_def_df.iloc[0][3:-2].index
print(element_names_row)
for i in range(len(mat_def_df)):
    # get material name
    mat_name = mat_def_df.iloc[i]['Material'] + " - C5"
    # get material id
    print(mat_name)
    mat_id = engine.execute(f"SELECT MaterialID FROM Materials WHERE Name = '{mat_name}'").fetchone()
    # get material composition
    mat_comp = mat_def_df.iloc[i][3:-2]
    # get element ids
    element_ids = [atomic_symbol_to_atomic_number(element) for element in mat_comp.index]
    # get weight fractions
    weight_fractions = mat_comp.values
    # create seperate entry for each elementID, weightFraction pair
    for element_id, weight_fraction in zip(element_ids, weight_fractions):
        # remove if weight_fraction == 0
        if weight_fraction == 0:
            continue
        mat_comp_df = mat_comp_df.append({'MaterialID': int(mat_id[0]), 'ElementID': int(element_id), 'WeightFraction': weight_fraction/100}, ignore_index=True)
mat_comp_df = mat_comp_df.astype({'MaterialID': 'int32', 'ElementID': 'int32', 'WeightFraction': 'float64'})
mat_comp_df
    
```

```python
# add df to sqlite3 database
mat_comp_df.to_sql('MaterialCompositions', con=engine, if_exists='append', index=False)
```
